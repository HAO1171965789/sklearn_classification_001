本案例数据集Car Evaluation来自UCI，建模前对数据进行了适当的预处理。从汽车的价格、质量、及舒适度这三个方面的相关数据（详见数据说明）出发，对汽车的客户满意度（简称为汽车满意度）进行分类预测。

数据一共包含6个特征，1个类别变量（即汽车满意度），共1728个样本点。

数据集链接：http://archive.ics.uci.edu/ml/datasets/Car+Evaluation

|特征|说明|
|-----:|-----:|
|buying|购买价格（1：low；2：med；3：high；4：vhigh）|
|maint|保养价格（1：low；2：med；3：high；4：vhigh）|
|doors|门的个数（1：2；2：3；3：4；4：5more）|
|persons|载客数（1：2；2：4；3：more）|
|lug_boot|车身的大小（1：small；2：med；3：big）|
|safety|安全等级（1：low；2：med；3：high）|
|car_acceptability|汽车满意度（1：unacc；0：acc）|

# 1. 准备

```python
# 读取数据
import pandas as pd
data = pd.read_csv('./Car.csv')

data.head()
```

```python
## 观察样本类别分布
data.car_acceptability.value_counts()
```

```python
## 类别分布条形图
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']

plt.figure(figsize=(8, 6))
data.car_acceptability.value_counts().plot(kind='bar', rot=360, title='类别数量分布条形图')
plt.xticks([1, 0], ['满意', '不满意'])
```

```python
## 数据集划分
from sklearn.model_selection import train_test_split

X = data.drop(['car_acceptability'], axis=1)
y = data['car_acceptability']

train_x, test_x, train_y, test_y = train_test_split(X, y, test_size = 0.2, random_state = 10)
```

<div id="2"></div>
# 2. 逻辑回归

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression

# 模型构建与拟合
clf = LogisticRegression(random_state=10)
clf.fit(train_x, train_y)
```

```python
# 模型预测
y_pred = clf.predict(test_x)
```

```python
# 分类正确率
print("分类正确率：",round(clf.score(test_x, test_y),4))
```

```python
# 构建显示主要分类指标的文本报告
print(classification_report(test_y, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns

## 设置正常显示中文
sns.set(font='SimHei')
## 绘制热力图
ax = sns.heatmap(confusion_matrix(test_y, y_pred), 
                 annot=True, fmt='d', 
                 xticklabels=["满意(0)","不满意(1)"], yticklabels=["满意(0)","不满意(1)"])
ax.set_ylabel('真实')
ax.set_xlabel('预测')
ax.set_title('混淆矩阵热力图')
```

```python
## 模型构建与拟合,平衡类别权重
clf = LogisticRegression(random_state=10, class_weight ='balanced')
clf.fit(train_x, train_y)
```

```python
# 模型预测
y_pred = clf.predict(test_x)
# 分类正确率
print("分类正确率：",round(clf.score(test_x, test_y),4))
# 构建显示主要分类指标的文本报告
print(classification_report(test_y, y_pred))
```

```python
## 绘制热力图
ax = sns.heatmap(confusion_matrix(test_y, y_pred), 
                 annot=True, fmt='d', 
                 xticklabels=["满意(0)","不满意(1)"], yticklabels=["满意(0)","不满意(1)"])
ax.set_ylabel('真实')
ax.set_xlabel('预测')
ax.set_title('混淆矩阵热力图')
```

```python
## 查看模型系数和截距项
pd.Series(clf.coef_.reshape(-1), index=X.columns).plot(kind='barh', title='系数条形图')
```

# 3. K近邻

```python
from sklearn.neighbors import KNeighborsClassifier
# 模型构建与拟合
neigh = KNeighborsClassifier()
neigh.fit(train_x, train_y)

# 模型预测
y_pred = neigh.predict(test_x)

# 分类正确率
print("分类正确率：",round(neigh.score(test_x, test_y),4))

# 分类报告
print(classification_report(test_y, y_pred))

# 绘制热力图
ax = sns.heatmap(confusion_matrix(test_y, y_pred), 
                 annot=True, fmt='d', 
                 xticklabels=["满意(0)","不满意(1)"], yticklabels=["满意(0)","不满意(1)"])
ax.set_ylabel('真实')
ax.set_xlabel('预测')
ax.set_title('混淆矩阵热力图')
```

```python
# 不同K值对分类正确率的影响
K_scores = [KNeighborsClassifier(n_neighbors=k).fit(train_x, train_y).score(test_x, test_y) for k in range(1,10)]

# 绘制折线图
import matplotlib.pyplot as plt
plt.plot(range(1, 10), K_scores, 'r-.', linewidth=3)
plt.xlabel('K')
plt.ylabel('正确率')
plt.title('不同K值下模型预测正确率')
```

```python
# 距离加权下不同K值对分类正确率的影响
K_scores_weights = [KNeighborsClassifier(n_neighbors=k,weights='distance').fit(train_x, train_y).score(test_x, test_y) for k in range(1,10)]

# 绘制折线图
import matplotlib.pyplot as plt
plt.plot(range(1, 10), K_scores_weights, 'g-.', linewidth=3)
plt.xlabel('K')
plt.ylabel('正确率')
plt.title('不同K值下模型预测正确率')
```

# 4. 决策树

```python
from sklearn.tree import DecisionTreeClassifier
# 模型构建与拟合
DF_model = DecisionTreeClassifier(random_state=10)
DF_model.fit(train_x, train_y)

# 模型预测
y_pred = DF_model.predict(test_x)

# 分类正确率
print("分类正确率：",round(DF_model.score(test_x, test_y),4))

# 分类报告
print(classification_report(test_y, y_pred))

# 绘制热力图
ax = sns.heatmap(confusion_matrix(test_y, y_pred), 
                 annot=True, fmt='d', 
                 xticklabels=["满意(0)","不满意(1)"], yticklabels=["满意(0)","不满意(1)"])
ax.set_ylabel('真实')
ax.set_xlabel('预测')
ax.set_title('混淆矩阵热力图')
```

```python
# 特征重要性
pd.Series(DF_model.feature_importances_, index=train_x.columns).sort_values().plot(kind='barh', title='特征重要性')
```

```python
# 限制树深度对分类正确率的影响
depth_grid = [1, 3, 5, 7, 9, None]
depth_scores = [DecisionTreeClassifier(random_state=10, max_depth=item).fit(train_x, train_y)
                .score(test_x, test_y) for item in depth_grid]

# 绘制折线图
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.plot(depth_scores, 'b-*', linewidth=1)
plt.xticks(range(6), [1, 3, 5, 7, 9, 'None'])
plt.xlabel('深度')
plt.ylabel('正确率')
plt.title('不同树深度下模型分类正确率')
```

```python
DF_model_3 = DecisionTreeClassifier(random_state=10, max_depth=3)
DF_model_3.fit(train_x, train_y)

# 决策树可视化
## 导入必要库
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image 
import pydotplus

## 输出图片到dot文件
export_graphviz(DF_model_3, out_file='tree.dot', 
                feature_names=train_x.columns,
                rounded=True, filled=True,
                class_names=['acc', 'unacc'])

## 使用dot文件构造图
graph= pydotplus.graph_from_dot_file('tree.dot')
Image(graph.create_png())
```

# 5. 朴素贝叶斯

## 5.1 Gaussian朴素贝叶斯

```python
## 载入算法类
from sklearn.naive_bayes import GaussianNB

# 模型构建与拟合
GNB= GaussianNB()
GNB.fit(train_x, train_y)

# 模型预测
y_pred = GNB.predict(test_x)

# 分类正确率
print("分类正确率：",round(GNB.score(test_x, test_y),4))

# 分类报告
print(classification_report(test_y, y_pred))

# 绘制热力图
ax = sns.heatmap(confusion_matrix(test_y, y_pred), 
                 annot=True, fmt='d', 
                 xticklabels=["满意(0)","不满意(1)"], yticklabels=["满意(0)","不满意(1)"])
ax.set_ylabel('真实')
ax.set_xlabel('预测')
ax.set_title('混淆矩阵热力图')
```

```python
## 查看先验概率和特征的均值方差
### 先验概率
print(GNB.class_prior_)

### 均值
print(GNB.theta_)

### 方差
print(GNB.sigma_)
```

```python
## 不同类别下特征均值条形图
pd.DataFrame(GNB.theta_.T, columns=['0类', '1类']).plot(kind='bar',rot=360, title='不同类别下特征均值条形图')
plt.ylim(0, 3.5)
plt.xticks(range(6), train_x.columns)
```

```python
## 不同类别下特征方差条形图
pd.DataFrame(GNB.sigma_.T, columns=['0类', '1类']).plot(kind='bar',rot=360, title='不同类别下特征方差条形图')
#plt.ylim(0, 2)
plt.xticks(range(6), train_x.columns)
```

# 6. 支持向量机

## 6.1 LinearSVC

```python
# 载入算法类
from sklearn.svm import LinearSVC

# 模型构建与拟合
lsvm = LinearSVC(random_state=10)
lsvm.fit(train_x, train_y)

# 模型预测
y_pred = lsvm.predict(test_x)

# 分类正确率
print("分类正确率：",round(lsvm.score(test_x, test_y),4))

# 分类报告
print(classification_report(test_y, y_pred))

# 绘制热力图
ax = sns.heatmap(confusion_matrix(test_y, y_pred), 
                 annot=True, fmt='d', 
                 xticklabels=["满意(0)","不满意(1)"], yticklabels=["满意(0)","不满意(1)"])
ax.set_ylabel('真实')
ax.set_xlabel('预测')
ax.set_title('混淆矩阵热力图')
```

```python
# 调整惩罚系数的取值
C_grid = [0.1, 0.3, 0.5, 0.7, 1, 2, 5, 10, 100]
C_scores = [LinearSVC(C=c, random_state=10).fit(train_x, train_y).score(test_x, test_y) 
            for c in C_grid]

# 绘制折线图
plt.figure(figsize=(8, 6))
plt.plot(C_scores, 'm-x', linewidth=2)
plt.xticks(range(9), C_grid)
plt.xlabel('惩罚系数')
plt.ylabel('正确率')
plt.title('不同惩罚系数下模型分类正确率')
```

```python
# 查看模型系数
pd.Series(lsvm.coef_.reshape(-1), index=train_x.columns).plot(kind='barh', title='系数条形图')
```

## 6.2 SVC

```python
# 载入算法类
from sklearn.svm import SVC

# 模型构建与拟合
ksvm = SVC(random_state=10)
ksvm.fit(train_x, train_y)

# 模型预测
y_pred = ksvm.predict(test_x)

# 分类正确率
print("分类正确率：",round(ksvm.score(test_x, test_y),4))

# 分类报告
print(classification_report(test_y, y_pred))

# 绘制热力图
ax = sns.heatmap(confusion_matrix(test_y, y_pred), 
                 annot=True, fmt='d', 
                 xticklabels=["满意(0)","不满意(1)"], yticklabels=["满意(0)","不满意(1)"])
ax.set_ylabel('真实')
ax.set_xlabel('预测')
ax.set_title('混淆矩阵热力图')
```

```python
# 调整核函数
kernel_grid = ['poly','sigmoid']
kernel_scores = [SVC(kernel=k, random_state=10).fit(train_x, train_y).score(test_x, test_y) 
                for k in kernel_grid]
print(pd.Series(kernel_scores, index=['poly', 'sigmoid']))
```

```python
# 调整惩罚系数的取值
C_grid = [0.1, 0.3, 0.5, 0.7, 1, 2, 5, 10, 100]
C_scores = [SVC(C=c, random_state=10).fit(train_x, train_y).score(test_x, test_y) 
            for c in C_grid]

# 绘制折线图
plt.figure(figsize=(8, 6))
plt.plot(C_scores, 'k-o', linewidth=2)
plt.xticks(range(9), C_grid)
plt.xlabel('惩罚系数')
plt.ylabel('正确率')
plt.title('不同惩罚系数下模型分类正确率')
```
